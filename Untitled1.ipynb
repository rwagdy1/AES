```python
from Crypto.Cipher import AES
from Crypto.Util.Padding import pad, unpad
from Crypto.Random import get_random_bytes
import binascii

# ══════════════════════════════════════════════════
#   AES Encryption & Decryption — CBC Mode
#   Key size: 16 bytes = AES-128 | 32 bytes = AES-256
# ══════════════════════════════════════════════════

# المفتاح: 16 bytes لـ AES-128 أو 32 bytes لـ AES-256
key = get_random_bytes(16)   # AES-128 (random key)
# key = get_random_bytes(32) # AES-256 (uncomment for stronger encryption)


def aes_encrypt(plaintext: str) -> bytes:
    """
    تشفير نص باستخدام AES-CBC
    - بتولّد IV عشوائي كل مرة (16 bytes)
    - بتعمل PKCS7 padding
    - بترجع: IV + Ciphertext
    """
    cipher = AES.new(key, AES.MODE_CBC)
    iv = cipher.iv                                        # 16 bytes random IV
    padded = pad(plaintext.encode('utf-8'), AES.block_size)
    ciphertext = cipher.encrypt(padded)
    return iv + ciphertext                                # IV + ciphertext معاً


def aes_decrypt(data: bytes) -> str:
    """
    فك تشفير البيانات
    - بتاخد أول 16 bytes كـ IV
    - بتفك الـ padding
    - بترجع النص الأصلي
    """
    iv         = data[:16]                                # أول 16 bytes هي الـ IV
    ciphertext = data[16:]
    cipher     = AES.new(key, AES.MODE_CBC, iv=iv)
    decrypted  = unpad(cipher.decrypt(ciphertext), AES.block_size)
    return decrypted.decode('utf-8')


# ══════════════════════════════════════════════════
#   تجربة الكود
# ══════════════════════════════════════════════════

if __name__ == "__main__":
    message = "Hello AES-128! هذه رسالة مشفرة."

    print("=" * 55)
    print("         AES Encryption & Decryption Demo")
    print("=" * 55)

    # Encryption
    encrypted = aes_encrypt(message)
    enc_hex   = binascii.hexlify(encrypted).decode()

    # Decryption
    decrypted = aes_decrypt(encrypted)

    print(f"Original  : {message}")
    print(f"Key (hex) : {binascii.hexlify(key).decode()}")
    print(f"Encrypted : {enc_hex}")
    print(f"Decrypted : {decrypted}")
    print("=" * 55)

    # تحقق من صحة النتيجة
    assert message == decrypted, "Decryption failed!"
    print("✓ Encryption & Decryption successful!")
```
